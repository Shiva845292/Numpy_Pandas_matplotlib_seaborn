```python
import pandas as pd
from datetime import datetime
```

```python
df=pd.read_csv("XM06235_Trade Book_28_01_2023.csv")
df.head()
```

```python
date_format=pd.to_datetime(df["Date"],errors='coerce',format='%d/%m/%Y',)
Date=date_format.drop_duplicates()
print(type(Date))
```

```python
df.head()
```

```python
df.groupby(by=["Date","Trade Type"]).agg({'Trade Value  (₹)':sum})
```

```python
count=df.groupby(by="Date").count().reset_index()
count=count.rename(columns={'Time':'Count'})
count=count["Count"]
count
```

```python
agg_df = df.groupby(by=['Date','Trade Type']).agg({'Trade Value  (₹)':sum}).reset_index()
sell_df=agg_df[agg_df['Trade Type']=='SELL'].rename(columns={'Trade Value (₹)':'total_sell'})
buy_df=agg_df[agg_df['Trade Type']=='BUY'].rename(columns={'Trade Value (₹)':'total_buy'})
sell_df=sell_df.drop(['Trade Type'], axis=1)
buy_df=buy_df.drop(['Trade Type'], axis=1)
final_df=pd.merge(sell_df,buy_df, on='Date')
final_df.rename(columns={"Date":'Date','Trade Value  (₹)_x':'Total Sell','Trade Value  (₹)_y':'Total Buy'},inplace=True)
final_df["Gross Profit or Loss"]=final_df['Total Sell']-final_df['Total Buy']
final_df
```

```python
Finall=pd.concat([final_df,count], axis=1, join='outer')
```

```python
Finall.to_excel('Tading_History_2023.xlsx')
```
