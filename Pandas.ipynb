```python
import pandas as pd
import requests
req= requests.get("https://api.github.com/repos/pandas-dev/pandas/issues")
data=req.json()
main=pd.DataFrame(columns=["id","node_id","url","name","color","default","description"])
for i in range(len(data)):
    lab=pd.DataFrame(data[i]["labels"])
    main=main.append(lab, ignore_index=True)
main
```

```python
datas=pd.DataFrame(data, columns=["comments_url","repository_url","events_url","labels_url"])
datas
```

```python
l=[datas,main]
need=pd.concat(l,axis=1)
need
```
