```python
l1=[]
for i in read["timeStamp"]:
    l1.append(i[8:10])
read["day"]=l1
read
```

```python
A=read[read["addr"].str.startswith('A',na=False)]
A.zip
```

```python
import datetime
from datetime import date 
Date=[]
date=""
for i in read["timeStamp"]:
    date=i[0:10]
    year, month, day  = date.split('-')     
    day_name = datetime.date(int(year), int(month), int(day)) 
    Date.append(day_name.strftime("%A"))
read["Day"]=Date
read
```

```python
import re
services=[]
for i in read["title"]:
    service=list(map(str,re.findall("([A-Za-z]+\:)",i)))
    for i in service:
#         print(i[:-1])
        services.append(i[:-1])
read["service"]=services
read
```
